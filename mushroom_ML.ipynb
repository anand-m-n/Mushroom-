Since, all the records (8124) for the feature 'veil-type' is same, it does not contribute anything to the dataset. Therefore, dropping the 'veil-type' feature. 

```python
data.drop(['veil-type'], axis=1, inplace=True)
```

```python
data.head()
```

#### Encoding the categorical features to numeric values

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
col = data.columns
for i in col:
    data[i]=encode.fit_transform(data[i])
```

```python
data
```

#### Seperating target and independent features

```python
X = data.drop(['class'], axis=1)
y = data['class']
```

```python
X
```

```python
y
```

#### Splitting into test and train set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_class = DecisionTreeClassifier()
dt_class.fit(x_train, y_train)
```

```python
y_pred_dt = dt_class.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred_dt)
```

```python
accuracy_score(y_test, y_pred_dt)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_R = LogisticRegression(max_iter=500, )
log_R.fit(x_train, y_train)
y_pred_lr =log_R.predict(x_test) 
```

```python
confusion_matrix(y_pred_lr,y_test)
```

```python
accuracy_score(y_pred_lr,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred_lr, y_test))
```

### SVM Classifier

```python
from sklearn.svm import SVC
svm = SVC(gamma='scale')
svm.fit(x_train, y_train)
y_pred_sc = svm.predict(x_test)
```

```python
accuracy_score(y_pred_sc, y_test)
```

```python
confusion_matrix(y_pred_sc, y_test)
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_t=knn.predict(x_test)
    print((accuracy_score(y_t,y_test)*100))
```

Value of K=1 or K=2 gives the best accuracy

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)
y_pred_knn=knn.predict(x_test)
print((accuracy_score(y_pred_knn,y_test)*100))
```

```python
confusion_matrix(y_pred_knn,y_test)
```

### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred_gnb = gnb.predict(x_test)
```

```python
accuracy_score(y_pred_gnb,y_test)
```

```python
confusion_matrix(y_pred_gnb, y_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
y_pred_bnb = bnb.predict(x_test)
```

```python
accuracy_score(y_pred_bnb,y_test)
```

```python
confusion_matrix(y_pred_bnb,y_test)
```
